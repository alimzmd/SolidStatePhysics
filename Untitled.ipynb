```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

file_path = 'BiSe_232deg_38p7nm_static_6.h5'

with h5py.File(file_path, 'r') as f:
    # 1. Load data and remove the dimension of size 1
    # Shape goes from (20, 1, 576, 864) to (20, 576, 864)
    data = np.squeeze(f['scan0/image'][()])

    # 2. Create a 4x5 grid of plots
    fig, axes = plt.subplots(4, 5, figsize=(20, 12))
    axes = axes.flatten() # Flatten 2D array of axes for easy looping

    # 3. Plot each frame
    for i in range(20):
        # Use a high-contrast colormap since your max value is only 11
        im = axes[i].imshow(data[i], cmap='hot', vmin=0, vmax=11)
        axes[i].set_title(f"Frame {i}")
        axes[i].axis('off') # Hide x/y numbers for a cleaner look

    # 4. Add a single colorbar for the whole figure
    plt.tight_layout()
    fig.colorbar(im, ax=axes.ravel().tolist(), fraction=0.02, pad=0.04)
    plt.show()
```

```python
import pandas as pd

file_path = 'BiSe_232deg_38p7nm_static_6.h5'
paths = ['scan0/image', 'scan0/lastimage', 'scan0/positions', 'scan0/temperatures']

for path in paths:
    # Read the dataset into a Series or DataFrame
    try:
        data = pd.read_hdf(file_path, key=path)
        
        print(f"--- Path: {path} ---")
        print(f"Shape: {data.shape}")
        
        # Pandas handles the printing of small vs large data automatically
        if data.size < 50:
            print(f"Values:\n{data}")
        else:
            print(f"Stats: Min={data.values.min()}, Max={data.values.max()}, Mean={data.values.mean():.2f}")
        print("-" * 30)
    except Exception as e:
        print(f"Could not read {path} with pandas: {e}")
```

```python
import h5py
import pandas as pd
import numpy as np

file_path = 'BiSe_232deg_38p7nm_static_6.h5'

with h5py.File(file_path, 'r') as f:
    # 1. Handle Multidimensional Data (Images)
    # We keep these as NumPy arrays because Pandas isn't meant for 2D/3D image grids
    img = f['scan0/image'][()]
    
    # 2. Handle Tabular Data (Positions/Temperatures)
    # We convert these into a Pandas DataFrame for easier manipulation
    df_coords = pd.DataFrame({
        'positions': f['scan0/positions'][()].flatten(),
        'temperatures': f['scan0/temperatures'][()].flatten()
    })

print("--- Image Stats ---")
print(f"Mean Intensity: {img.mean():.2f}")

print("\n--- Scalar Data Summary ---")
print(df_coords.describe()) # Pandas gives you a full statistical breakdown instantly
```
