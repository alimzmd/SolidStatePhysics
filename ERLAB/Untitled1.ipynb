```python
# Assuming your script is named 'my_arpes_loader.py' and contains the function.
# You must specify the path to your data folder.
import data
data_folder = "C:/Users/mdram/SolidStatePhysics1/dataverse_files"

# The function returns a single object
arpes_data = load_ses_spectra(data_folder)
```

```python
import matplotlib.pyplot as plt

# Take a slice at the center of the ky dimension (ky = 0)
# 'method='nearest'' finds the closest available value
center_ky_slice = arpes_data.sel(ky=0, method='nearest')

# Plot the intensity map
plt.figure(figsize=(8, 6))
center_ky_slice.plot.pcolormesh(x='kx', y='energy')
plt.title("ARPES Band Map (Intensity vs. Energy and $k_x$)")
plt.xlabel("$k_x$ (Å⁻¹)")
plt.ylabel("Energy (eV)")
plt.show()
```

```python
# Take a slice at the Fermi level (Energy = 0 eV)
# If your energy axis does not go to 0, select the closest value
fermi_level_slice = arpes_data.sel(energy=0, method='nearest')

# Plot the Fermi surface map
plt.figure(figsize=(8, 6))
fermi_level_slice.plot.pcolormesh(x='kx', y='ky')
plt.title("Fermi Surface Map (Intensity vs. $k_x$ and $k_y$)")
plt.xlabel("$k_x$ (Å⁻¹)")
plt.ylabel("$k_y$ (Å⁻¹)")
plt.show()
```

```python
# Select a single point at kx=0 and ky=0
edc_cut = arpes_data.sel(kx=0, ky=0, method='nearest')

# Plot the 1D intensity data
plt.figure(figsize=(6, 4))
edc_cut.plot(x='energy')
plt.title("Energy Distribution Curve at $k_x=0, k_y=0$")
plt.xlabel("Energy (eV)")
plt.ylabel("Intensity (a.u.)")
plt.show()
```
