```python
import numpy as np
from matplotlib import rc
from scipy.special import factorial
import matplotlib.pyplot as plt

# --- CONFIGURATION ---
# Set usetex to False to avoid "latex not found" errors. 
# Matplotlib will use its internal "mathtext" parser instead.
rc('font', **{'family': 'serif', 'size': 12})
rc('text', usetex=False)

# PLOT_PROB=False plots the wavefunction, psi; PLOT_PROB=True plots |psi|^2
PLOT_PROB = False
VMAX = 6            # Max vibrational quantum number
QPAD_FRAC = 1.3     # Padding for the x-axis
SCALING = 0.6       # Scale wavefunctions to prevent overlap
COLOUR1 = (0.6196, 0.0039, 0.2588, 1.0) # Positive phase
COLOUR2 = (0.3686, 0.3098, 0.6353, 1.0) # Negative phase

# --- FUNCTIONS ---
# Normalization: N = [1 / sqrt(sqrt(pi) * 2^v * v!)]
N = lambda v: 1./np.sqrt(np.sqrt(np.pi) * 2**v * factorial(v))

# Energy levels: E = (v + 1/2)
get_E = lambda v: v + 0.5

def make_Hr():
    """Return a list of np.poly1d objects representing Hermite polynomials."""
    Hr = [None] * (VMAX + 1)
    Hr[0] = np.poly1d([1.,])
    Hr[1] = np.poly1d([2., 0.])
    for v in range(2, VMAX+1):
        Hr[v] = Hr[1]*Hr[v-1] - 2*(v-1)*Hr[v-2]
    return Hr

def get_psi(v, q):
    """Return the harmonic oscillator wavefunction for level v on grid q."""
    return N(v) * Hr[v](q) * np.exp(-q*q / 2.)

def get_turning_points(v):
    """Return the classical turning points where E = V(q)."""
    qmax = np.sqrt(2. * get_E(v))
    return -qmax, qmax

def plot_func(ax, q_grid, f, scaling=1, yoffset=0):
    """Plot f*scaling with offset yoffset and color-filled phases."""
    ax.plot(q_grid, f*scaling + yoffset, color=COLOUR1, lw=1)
    ax.fill_between(q_grid, f*scaling + yoffset, yoffset, where=(f > 0),
                    color=COLOUR1, alpha=0.4)
    ax.fill_between(q_grid, f*scaling + yoffset, yoffset, where=(f < 0),
                    color=COLOUR2, alpha=0.4)

# --- MAIN PLOTTING ---
Hr = make_Hr()
fig, ax = plt.subplots(figsize=(8, 10))

# Set the spatial grid based on the highest energy level's turning points
qmin_limit, qmax_limit = get_turning_points(VMAX)
xmin, xmax = QPAD_FRAC * qmin_limit, QPAD_FRAC * qmax_limit
q_plot = np.linspace(xmin, xmax, 500)

# Plot the Potential Well: V(q) = 0.5 * q^2
ax.plot(q_plot, 0.5 * q_plot**2, color='black', linewidth=2, label='Potential $V(q)$')

# Iterate through quantum levels
for v in range(VMAX + 1):
    psi_v = get_psi(v, q_plot)
    E_v = get_E(v)
    
    if PLOT_PROB:
        plot_func(ax, q_plot, psi_v**2, scaling=SCALING*1.5, yoffset=E_v)
        ylabel = r'$|\psi_v(q)|^2$'
    else:
        plot_func(ax, q_plot, psi_v, scaling=SCALING, yoffset=E_v)
        ylabel = r'$\psi_v(q)$'
    
    # Energy labels (Right side)
    ax.text(xmax + 0.1, E_v, r'$\frac{%d}{2}\hbar\omega$' % (2*v + 1), 
            va='center', fontsize=12)
    
    # Quantum number labels (Left side)
    ax.text(xmin - 0.1, E_v, r'$v=%d$' % v, va='center', ha='right', fontsize=12)

# Formatting
ymax = get_E(VMAX) + 1.0
ax.set_xlim(xmin - 0.5, xmax + 1.0)
ax.set_ylim(0, ymax)
ax.set_xlabel('$q$ (Dimensionless displacement)', fontsize=14)
ax.set_title(ylabel + ' for the Quantum Harmonic Oscillator', fontsize=16, pad=20)

# Clean up axes
ax.spines['left'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_yticks([]) # Hide Y ticks as levels are labeled individually

plt.tight_layout()
plt.savefig('2')
plt.show()
```

```python
import numpy as np
from scipy.special import factorial
import matplotlib.pyplot as plt

VMAX = 6
QPAD_FRAC = 1.3
SCALING = 0.7
COLOUR1 = (0.6196, 0.0039, 0.2588, 1.0)
COLOUR2 = (0.3686, 0.3098, 0.6353, 1.0)

N = lambda v: 1./np.sqrt(np.sqrt(np.pi) * 2**v * factorial(v))
get_E = lambda v: v + 0.5

def make_Hr():
    Hr = [np.poly1d([1.]), np.poly1d([2., 0.])]
    for v in range(2, VMAX + 1):
        Hr.append(np.poly1d([2., 0.]) * Hr[v-1] - 2 * (v-1) * Hr[v-2])
    return Hr

def get_psi(v, q, Hr):
    return N(v) * Hr[v](q) * np.exp(-q**2 / 2.)

def plot_harmonic_oscillator(plot_prob=False):
    Hr = make_Hr()
    fig, ax = plt.subplots(figsize=(7, 9))
    
    # Calculate limits based on highest state
    qmax_tp = np.sqrt(2. * get_E(VMAX))
    limit = QPAD_FRAC * qmax_tp
    q = np.linspace(-limit, limit, 500)
    
    ax.plot(q, q**2 / 2, color='k', linewidth=1.5)

    for v in range(VMAX + 1):
        psi_v = get_psi(v, q, Hr)
        E_v = get_E(v)
        f = psi_v**2 if plot_prob else psi_v
        s = SCALING * 1.5 if plot_prob else SCALING
        
        y_offset = f * s + E_v
        ax.plot(q, y_offset, color=COLOUR1, lw=1)
        ax.fill_between(q, y_offset, E_v, where=(f > 0), color=COLOUR1, alpha=0.5)
        ax.fill_between(q, y_offset, E_v, where=(f < 0), color=COLOUR2, alpha=0.5)
        
        # Consistent label placement
        ax.text(limit + 0.1, E_v, r'$\frac{%d}{2}\hbar\omega$' % (2*v+1), va='center')
        ax.text(-limit - 0.1, E_v, f'$v={v}$', va='center', ha='right')

    # FIX: Ensure the spine is exactly at 0 and limits are balanced
    ax.spines['left'].set_position(('data', 0)) 
    ax.set_xlim(-limit - 1.0, limit + 1.0) # Balanced padding for labels
    
    ax.set_xlabel('$q$')
    ax.xaxis.set_label_coords(0.5, -0.05) # Center the 'q' label
    ax.set_title(r'$|\psi(q)|^2$' if plot_prob else r'$\psi(q)$', pad=20)
    ax.set_ylim(0, VMAX + 1)
    ax.set_yticks([])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.tight_layout()

# Run both
plot_harmonic_oscillator(plot_prob=False)
plot_harmonic_oscillator(plot_prob=True)
plt.show()
```
