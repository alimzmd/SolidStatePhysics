```python
# No import needed as math library functions aren't used
mass = 5  # kg
radius = 0.2  # m
velocity = 3  # m/s

# 1. Calculate translational kinetic energy
ke_trans = 0.5 * mass * velocity**2
print(f"Translational KE: {ke_trans} Joules")

# 2. Calculate moment of inertia
I = 0.5 * mass * radius**2
print(f"Moment of Inertia: {I} kg*m^2")

# 3. Calculate angular velocity
omega = velocity / radius
print(f"Angular Velocity: {omega} rad/s")

# 4. Calculate rotational kinetic energy
ke_rot = 0.5 * I * omega**2
print(f"Rotational KE: {ke_rot} Joules")

# 5. Calculate total kinetic energy
total_ke = ke_trans + ke_rot
print(f"Total KE: {total_ke} Joules")
```

```python
import math

def rotational_kinetic_energy(m, r, v):
    # Moment of inertia for a solid cylinder
    I = 0.5 * m * r**2
    # Angular velocity
    omega = v / r
    
    ke_trans = 0.5 * m * v**2
    ke_rot = 0.5 * I * omega**2
    
    total_ke = ke_trans + ke_rot
    return total_ke

mass = 5  # kg
radius = 0.2  # m
velocity = 3  # m/s

total_energy = rotational_kinetic_energy(mass, radius, velocity)
print(total_energy)
```

```python
print( f"total_kinetic_energy: {total_energy:.2f} J")
```
